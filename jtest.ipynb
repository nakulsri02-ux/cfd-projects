```python
import numpy as np
import matplotlib.pyplot as plt

# Test NumPy
print("NumPy version:", np.__version__)
print("NumPy is working! ✓")

# Create some data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Plot it
plt.figure(figsize=(8, 4))
plt.plot(x, y, 'b-', linewidth=2)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('My First CFD Plot - Testing Setup')
plt.grid(True, alpha=0.3)
plt.show()

print("\nMatplotlib is working! ✓")
print("You're ready to build CFD solvers!")
```
